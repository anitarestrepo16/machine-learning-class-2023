```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import statsmodels.formula.api as smf
import seaborn as sns
```

```python
# Load the dataset
boston = pd.read_csv('Boston.csv')
boston.head()
```

```python
sns.pairplot(boston)
```

```python
boston.plot(y='CRIM', kind='hist', title='Distribution of Crime Rates')
boston.plot(y='TAX', kind='hist', title='Distribution of Tax Rates')
boston.plot(y='PTRATIO', kind='hist', title='Distribution of Pupil-Teacher Ratios')
```

```python
names = ['not_bound', 'bound']
values = [boston.CHAS.value_counts()[0], boston.CHAS.value_counts()[1]]
print(values)
plt.plot()
plt.bar(names, values)
```
